```python
#6.
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(['cat', 'dog', 'mouse','elephant'])
print(le.fit_transform(['cat', 'dog', 'mouse','elephant']))
print(le.transform(['elephant'])) 
```

```python
#7.
from sklearn.impute import SimpleImputer
import numpy as np

X = np.array([[1, 2], [np.nan, 3], [7, 6]])

imp = SimpleImputer(strategy='mean')
imp.fit(X)
print(imp.fit_transform(X))
print(imp.transform([[np.nan, 3]])) 
```
